```python
# Create a dataframe for AIS data, with fields for MMSI, YMD_HMS, SOG, COG, Latitude, Longitude, and Vessel Name

import pandas as pd
import numpy as np
import random
import datetime
import time

df = pd.DataFrame(columns=['MMSI', 'YMD_HMS', 'SOG', 'COG', 'Latitude', 'Longitude', 'Vessel Name'])

# Populate the dataframe with random data
for i in range(1000):
    random_seconds = random.randint(0, 3600000)
    date_time = datetime.datetime.now() - datetime.timedelta(seconds=random_seconds)
    df.loc[i] = [random.randint(100000000,999999999), 
                 date_time.strftime('%Y-%m-%d %H:%M:%S'), 
                 random.randint(0,50), 
                 random.randint(0,360), 
                 random.uniform(45,55),
                 random.uniform(-130,-120), 
                 'Vessel ' + str(random.randint(1,20))]
```
